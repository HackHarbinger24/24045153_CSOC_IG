##### Above: N=Number Of Features  M=Number Of Examples

```python
def ceil(x):
    i = int(x)
    return i if x==i else i+(1 if x>0 else 0)
```

```python
def dot_product(u,v):
    if len(u)!=len(v):
        raise ValueError('The dimension of u and v must be equal')
    return sum(a*b for a,b in zip(u,v))
```

```python
def sgn(x):
    if x<0:
        return -1
    elif x>0:
        return 1
    else:
        return 0
```

#### Mean Absolute Error(MAE):

```python
def compute_cost_MAE(x,y,w,b):
    cost=0
    for i in range(m):
        f_wb=dot_product(w,x[i])+b
        loss=abs(f_wb-y[i])
        cost=cost+loss
    cost=cost/m
    return cost
```

```python
def compute_gradient_MAE(x,y,w,b):
    dj_dw=[0 for _ in range(n)]
    dj_db=0
    for i in range(n):
        dj_dw_f=0
        for j in range(m):
            f_wb=dot_product(w,x[j])+b
            dj_dw_i=sgn(f_wb-y[j])*x[j][i]
            dj_dw_f=dj_dw_f+dj_dw_i
        dj_dw_f=dj_dw_f/m
        dj_dw[i]=dj_dw_f

    for i in range(m):
        f_wb=dot_product(w,x[i])+b
        dj_db_i=sgn(f_wb-y[i])
        dj_db=dj_db+dj_db_i
    dj_db=dj_db/m
    return dj_dw,dj_db
```

#### Root Mean Squared Error (RMSE):

```python
def compute_cost_RMSE(x,y,w,b):
    cost=0
    for i in range(m):
        f_wb=dot_product(w,x[i])+b
        loss=(f_wb-y[i])**2
        cost=cost+loss
    cost=cost/m
    cost=cost**0.5
    return cost
```

```python
def compute_gradient_RMSE(x,y,w,b):
    rmse=compute_cost_RMSE(x,y,w,b)
    dj_dw=[0 for _ in range(n)]
    dj_db=0
    for i in range(n):
        dj_dw_f=0
        for j in range(m):
            f_wb=dot_product(w,x[j])+b
            dj_dw_i=(f_wb-y[j])*x[j][i]
            dj_dw_f=dj_dw_f+dj_dw_i
        dj_dw_f=dj_dw_f/(rmse*m)
        dj_dw[i]=dj_dw_f

    for i in range(m):
        f_wb=dot_product(w,x[i])+b
        dj_db_i=(f_wb-y[i])
        dj_db=dj_db+dj_db_i
    dj_db=dj_db/(rmse*m)
    return dj_dw, dj_db
```

#### R-Squared,R^2

```python
def sst(y):
    y_=0
    sst=0
    for i in range(m):
        y_=y_+y[i]
    y_=y_/m
    for i in range(m):
        SST=(y[i]-y_)**2
        sst=sst+SST
    return sst
```

```python
def compute_cost_R_Square(x,y,w,b,sst):
    cost=0
    ssr=0
    for i in range(m):
        f_wb=dot_product(w,x[i])+b
        SSR=(f_wb-y[i])**2
        ssr=ssr+SSR
    cost=1-(ssr/sst)
    return cost
```

```python
def compute_gradient_R_Square(x,y,w,b,sst):
    dj_dw=[0 for _ in range(n)]
    dj_db=0
    for i in range(n):
        dj_dw_f=0
        for j in range(m):
            f_wb=dot_product(w,x[j])+b
            dj_dw_i=(f_wb-y[j])*x[j][i]
            dj_dw_f=dj_dw_f+dj_dw_i
        dj_dw_f=(dj_dw_f*2)/sst
        dj_dw[i]=dj_dw_f

    for i in range(m):
        f_wb=dot_product(w,x[i])+b
        dj_db_i=(f_wb-y[i])
        dj_db=dj_db+dj_db_i
    dj_db=(dj_db*2)/sst
    return dj_dw, dj_db
```

```python
def gradient_descent(x,y,win,bin,alpha,num_iter):
    w_M,w_R,w_R_S=win,win,win
    b_M,b_R,b_R_S=bin,bin,bin
    a=alpha
    for i in range(num_iter):
        "First compute gradient descent for all Loss Functions"
        "MAE"
        dj_dw_MAE,dj_db_MAE=compute_gradient_MAE(x,y,w_M,b_M)
        "RMSE"
        dj_dw_RMSE,dj_db_RMSE=compute_gradient_RMSE(x,y,w_R,b_R)
        "R_Square"
        s=sst(y)
        dj_dw_R_Square,dj_db_R_Square=compute_gradient_R_Square(x,y,w_R_S,b_R_S,s)
        "Update weights and biases"
        "MAE"
        for k in range(n):
            w_M[k]=w_M[k]-a*dj_dw_MAE[k]
            w_R[k]=w_R[k]-a*dj_dw_RMSE[k]
            w_R_S[k]=w_R_S[k]-a*dj_dw_R_Square[k]
        b_M=b_M-a*dj_db_MAE
        b_R=b_R-a*dj_db_RMSE
        b_R_S=b_R_S-a*dj_db_R_Square
        "saving the cost and w,b"
        if i<100000:
            "Calculating cost and storing for graph"
            cost_MAE=compute_cost_MAE(x,y,w_M,b_M)
            cost_RMSE=compute_cost_RMSE(x,y,w_R,b_R)
            cost_R_Square=compute_cost_R_Square(x,y,w_R_S,b_R_S,s)
            "print cost,weights and biases"
            if i%ceil(num_iter/10)==0:
                print(f"MAE: Iteration {i:4} : Cost {cost_MAE}")
                print(f"RMSE: Iteration {i:4} : Cost {cost_RMSE}")
                print(f"R_Square: Iteration {i:4} : Cost {cost_R_Square}")
    return w_M,w_R,w_R_S,b_M,b_R,b_R_S
```

```python
"Set Initial Weights and Biases"
bin=0
win=[0 for _ in range(n)]
alpha=8.0e-2
num_iter=1500
"Run Gradient Descent"
w_f_M,w_f_R,w_f_R_S,b_f_M,b_f_R,b_f_R_S=gradient_descent(x_train_list,y_train_list,win,bin,alpha,num_iter)
print(f"Final weight for MAE : {w_f_M}")
print(f"Final weight for MAE : {w_f_R}")
print(f"Final weight for MAE : {w_f_R_S}")
print(f"Final bias for MAE: {b_f_M}")
print(f"Final bias for MAE: {b_f_R}")
print(f"Final bias for MAE: {b_f_R_S}")
```

#### Mean Absolute Error:

```python
def compute_cost_MAE_numpy(x,y,w,b):
    m=len(x)
    cost=0
    for i in range(m):
        f_wb=np.dot(w,x[i])+b
        loss=abs(f_wb-y[i])
        cost=cost+loss
    cost=cost/m
    return cost
```

```python
def compute_gradient_MAE_numpy(x,y,w,b):
    dj_dw=np.zeros(n)
    dj_db=0
    for i in range(n):
        dj_dw_f=0
        for j in range(m):
            f_wb=np.dot(w,x[j])+b
            dj_dw_i=sgn(f_wb-y[j])*x[j][i]
            dj_dw_f=dj_dw_f+dj_dw_i
        dj_dw_f=dj_dw_f/m
        dj_dw[i]=dj_dw_f

    for i in range(m):
        f_wb=np.dot(w,x[i])+b
        dj_db_i=sgn(f_wb-y[i])
        dj_db=dj_db+dj_db_i
    dj_db=dj_db/m
    return dj_dw,dj_db
```

#### Root Mean Square Error:

```python
def compute_cost_RMSE_numpy(x,y,w,b):
    m=len(x)
    cost=0
    for i in range(m):
        f_wb=np.dot(w,x[i])+b
        loss=(f_wb-y[i])**2
        cost=cost+loss
    cost=cost/m
    cost=cost**0.5
    return cost
```

```python
def compute_gradient_RMSE_numpy(x,y,w,b):
    [m,n]=x.shape
    rmse=compute_cost_RMSE_numpy(x,y,w,b)
    dj_dw=np.zeros(n)
    dj_db=0
    for i in range(n):
        dj_dw_f=0
        for j in range(m):
            f_wb=np.dot(w,x[j])+b
            dj_dw_i=(f_wb-y[j])*x[j][i]
            dj_dw_f=dj_dw_f+dj_dw_i
        dj_dw_f=dj_dw_f/(rmse*m)
        dj_dw[i]=dj_dw_f

    for i in range(m):
        f_wb=np.dot(w,x[i])+b
        dj_db_i=(f_wb-y[i])
        dj_db=dj_db+dj_db_i
    dj_db=dj_db/(rmse*m)
    return dj_dw,dj_db
```

#### R_Square:

```python
def compute_cost_R_Square_numpy(x,y,w,b):
    m=len(x)
    cost=0
    sst=0
    ssr=0
    y_=np.mean(y)
    for i in range(m):
        f_wb=np.dot(w,x[i])+b
        SSR=(f_wb-y[i])**2
        SST=(y[i]-y_)**2
        ssr=ssr+SSR
        sst=sst+SST
    cost=1-ssr/sst
    return cost
```

```python
def compute_gradient_R_Square_numpy(x,y,w,b):
    [m,n]=x.shape
    y_=np.mean(y)
    sst=0
    for i in range(m):
        SST=(y[i]-y_)**2
        sst=sst+SST
    dj_dw=np.zeros(n)
    dj_db=0
    for i in range(n):
        dj_dw_f=0
        for j in range(m):
            f_wb=np.dot(w,x[j])+b
            dj_dw_i=(f_wb-y[j])*x[j][i]
            dj_dw_f=dj_dw_f+dj_dw_i
        dj_dw_f=(dj_dw_f*2)/sst
        dj_dw[i]=dj_dw_f

    for i in range(m):
        f_wb=np.dot(w,x[i])+b
        dj_db_i=(f_wb-y[i])
        dj_db=dj_db+dj_db_i
    dj_db=(dj_db*2)/sst
    return dj_dw, dj_db
```

```python
def gradient_descent(x,y,win,bin,alpha,num_iter):
    w_M,w_R,w_R_S=win,win,win
    b_M,b_R,b_R_S=bin,bin,bin
    a=alpha
    for i in range(num_iter):
        "First compute gradient descent for all Loss Functions"
        "MAE"
        dj_dw_MAE,dj_db_MAE=compute_gradient_MAE_numpy(x,y,w_M,b_M)
        "RMSE"
        dj_dw_RMSE,dj_db_RMSE=compute_gradient_RMSE_numpy(x,y,w_R,b_R)
        "R_Square"
        dj_dw_R_Square,dj_db_R_Square=compute_gradient_R_Square_numpy(x,y,w_R_S,b_R_S)
        "Update weights and biases"
        "MAE"
        w_M=w_M+a*dj_dw_MAE
        w_R=w_R+a*dj_dw_RMSE
        w_R_S=w_R_S+a*dj_dw_R_Square
        b_M=b_M-a*dj_db_MAE
        b_R=b_R-a*dj_db_RMSE
        b_R_S=b_R_S-a*dj_db_R_Square
        "saving the cost and w,b"
        if i<100000:
            "Calculating cost and storing for graph"
            cost_MAE=compute_cost_MAE_numpy(x,y,w_M,b_M)
            cost_RMSE=compute_cost_RMSE_numpy(x,y,w_R,b_R)
            cost_R_Square=compute_cost_R_Square_numpy(x,y,w_R_S,b_R_S)
            "print cost,weights and biases"
            if i%ceil(num_iter/10)==0:
                print(f"MAE: Iteration {i:4} : Cost {cost_MAE}")
                print(f"RMSE: Iteration {i:4} : Cost {cost_RMSE}")
                print(f"R_Square: Iteration {i:4} : Cost {cost_R_Square}")
    return w_M,w_R,w_R_S,b_M,b_R,b_R_S
```

```python
"Set Initial Weights and Biases"
bin=0
win=np.zeros(n)
alpha=8.0e-2
num_iter=1000
"Run Gradient Descent"
w_f_M_Num,w_f_R_Num,w_f_R_S_Num,b_f_M_Num,b_f_R_Num,b_f_R_S_Num=gradient_descent(x_train_numpy,y_train_numpy,win,bin,alpha,num_iter)
print(f"Final weight for MAE : {w_f_M_Num}")
print(f"Final weight for MAE : {w_f_R_Num}")
print(f"Final weight for MAE : {w_f_R_S_Num}")
print(f"Final bias for MAE: {b_f_M_Num}")
print(f"Final bias for MAE: {b_f_R_Num}")
print(f"Final bias for MAE: {b_f_R_S_Num}")
```
